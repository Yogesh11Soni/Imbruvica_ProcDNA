```python
import pandas as pd
from ntscraper import Nitter
```

```python
scraper = Nitter()
```

```python
def get_tweets():
    terms = ["Imbruvica"]
    tweets = scraper.get_tweets(terms, mode = 'term', number=1000)
    final_tweets = []
    for x in tweets['tweets']:
        data = [x['user']['name'], x['user']['username'],x['link'], x['text'],x['stats']['likes'],x['stats']['retweets'], x['stats']['comments']]
        final_tweets.append(data)
    data= pd.DataFrame(final_tweets, columns =['Name','username','twitter_link','text','likes','retweets', 'replies'])
    return data
```

```python
data = get_tweets()
```

```python
data.to_csv('imbruvica.csv')
```

## Removing Rows having language other than english

```python
from langdetect import detect
def detect_language(text):
    try:
        return detect(text)
    except:
        return 'unknown'

def filter_non_english(df):
    df['language'] = df['text'].apply(detect_language)
    english_only_df = df[df['language'] == 'en']  # Filter for English rows
    english_only_df = english_only_df.drop(columns=['language'])  # Optionally drop the 'language' column
    return english_only_df

df = filter_non_english(data)
df
```

# analizing sentiments of the tweets

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

physician_df = pd.DataFrame(df)

# Initialize VaderSentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Function to get sentiment score
def get_sentiment_score(sentence):
    return analyzer.polarity_scores(sentence)['compound']

# Apply sentiment analysis to the 'text' column
physician_df['sentiment_score'] = physician_df['text'].apply(get_sentiment_score)

# Categorize sentiments
def categorize_sentiment(score):
    if score >= 0.05:
        return 'Positive'
    elif score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

physician_df['sentiment'] = physician_df['sentiment_score'].apply(categorize_sentiment)

# Calculate percentage of positive, negative, and neutral tweets
sentiment_counts = physician_df['sentiment'].value_counts(normalize=True) * 100

print("Percentage of tweets categorized as:")
print(f"Positive: {sentiment_counts.get('Positive', 0):.2f}%")
print(f"Negative: {sentiment_counts.get('Negative', 0):.2f}%")
print(f"Neutral: {sentiment_counts.get('Neutral', 0):.2f}%")
```

```python
physician_df = pd.DataFrame(df)

# Function to check if a tweet is from a physician based on keywords
def is_physician_tweet(text):
    keywords = ['doctor', 'physician', 'medtwitter', 'healthcare', 'clinical research', 'public health']  
    text_lower = text.lower()
    return any(keyword in text_lower for keyword in keywords)

# Filter tweets made by physicians
physician_tweets = physician_df[physician_df['text'].apply(is_physician_tweet)]

# Initialize VaderSentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Function to get sentiment score
def get_sentiment_score(sentence):
    return analyzer.polarity_scores(sentence)['compound']

# Apply sentiment analysis to the physician tweets
physician_tweets['sentiment_score'] = physician_tweets['text'].apply(get_sentiment_score)

# Categorize sentiments
def categorize_sentiment(score):
    if score >= 0.05:
        return 'Positive'
    elif score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

physician_tweets['sentiment'] = physician_tweets['sentiment_score'].apply(categorize_sentiment)

# Calculate percentage of positive, negative, and neutral tweets made by physicians
sentiment_counts_physicians = physician_tweets['sentiment'].value_counts(normalize=True) * 100

print("Sentiment analysis results for tweets made by physicians:")
print(f"Positive: {sentiment_counts_physicians.get('Positive', 0):.2f}%")
print(f"Negative: {sentiment_counts_physicians.get('Negative', 0):.2f}%")
print(f"Neutral: {sentiment_counts_physicians.get('Neutral', 0):.2f}%")
```

# Data Cleaning

## converting followers data into integers

```python
def convert_followers(followers):
    if followers is None:
        return 0  # Replace None with 0 or handle as appropriate
    followers = followers.replace(',', '') if isinstance(followers, str) else followers
    if isinstance(followers, str):
        followers_lower = followers.lower()
        if followers_lower.endswith('k'):
            return int(float(followers_lower.replace('k', '')) * 1000)
        elif followers_lower.endswith('m'):
            return int(float(followers_lower.replace('m', '')) * 1000000)
        else:
            return int(followers)
    else:
        return int(followers)


physician_tweets['followers'] = physician_tweets['followers'].apply(convert_followers)
```
